```python
import random
from typing import Optional, Tuple
import matplotlib.pyplot as plt
import networkx as nx
from IPython.display import display, clear_output
import time

class Node:
    def __init__(self, x: int, y: Optional[float] = None):
        self.x = x
        self.y = y if y is not None else random.random()
        self.left: Optional['Node'] = None
        self.right: Optional['Node'] = None

    def __str__(self):
        return f"Node(x={self.x}, y={self.y:.2f})"

class Treap:
    def __init__(self):
        self.root: Optional[Node] = None

    def split(self, node: Optional[Node], x: int) -> Tuple[Optional[Node], Optional[Node]]:
        if node is None:
            return None, None

        if node.x < x:
            left, right = self.split(node.right, x)
            node.right = left
            return node, right
        else:
            left, right = self.split(node.left, x)
            node.left = right
            return left, node

    def merge(self, left: Optional[Node], right: Optional[Node]) -> Optional[Node]:
        if left is None:
            return right
        if right is None:
            return left

        if left.y > right.y:
            left.right = self.merge(left.right, right)
            return left
        else:
            right.left = self.merge(left, right.left)
            return right

    def insert(self, x: int, y: Optional[float] = None):
        new_node = Node(x, y)
        if self.root is None:
            self.root = new_node
            return

        current = self.root
        parent = None
        while current is not None and current.y >= new_node.y:
            parent = current
            if new_node.x < current.x:
                current = current.left
            else:
                current = current.right

        if current is not None:
            left, right = self.split(current, new_node.x)
            new_node.left = left
            new_node.right = right
        else:
            left = right = None

        if parent is None:
            self.root = new_node
        elif new_node.x < parent.x:
            parent.left = new_node
        else:
            parent.right = new_node

    def remove(self, x: int):
        def _remove(node: Optional[Node], x: int) -> Optional[Node]:
            if node is None:
                return None

            if x < node.x:
                node.left = _remove(node.left, x)
            elif x > node.x:
                node.right = _remove(node.right, x)
            else:
                return self.merge(node.left, node.right)
            return node

        self.root = _remove(self.root, x)

    def contains(self, x: int) -> bool:
        current = self.root
        while current is not None:
            if x == current.x:
                return True
            elif x < current.x:
                current = current.left
            else:
                current = current.right
        return False

    def inorder(self) -> list:
        result = []
        self._inorder(self.root, result)
        return result

    def _inorder(self, node: Optional[Node], result: list):
        if node is not None:
            self._inorder(node.left, result)
            result.append((node.x, node.y))
            self._inorder(node.right, result)

    def __str__(self):
        return str([(x, f"{y:.2f}") for x, y in self.inorder()])

class TreapVisualizer:
    def __init__(self):
        self.fig, self.ax = plt.subplots(figsize=(10, 6))
        self.pos = {}

    def _build_graph(self, node, graph=None, x=0, y=0, dx=1):
        if graph is None:
            graph = nx.DiGraph()

        if node is None:
            return graph, x

        self.pos[node.x] = (x, -y)
        graph.add_node(node.x, y=f"{node.y:.2f}")

        left_graph, x = self._build_graph(node.left, graph, x-dx, y+1, dx/2)
        right_graph, x = self._build_graph(node.right, left_graph, x+dx, y+1, dx/2)

        if node.left is not None:
            right_graph.add_edge(node.x, node.left.x)
        if node.right is not None:
            right_graph.add_edge(node.x, node.right.x)

        return right_graph, x

    def draw_treap(self, treap, title="", highlight=None):
        self.ax.clear()

        if treap.root is None:
            self.ax.text(0.5, 0.5, "Treap is empty", ha='center', fontsize=14)
            plt.axis('off')
            display(self.fig)
            clear_output(wait=True)
            return

        G, _ = self._build_graph(treap.root)

        node_colors = ['lightgreen' if n == highlight else 'skyblue' for n in G.nodes()]
        nx.draw_networkx_nodes(G, self.pos, ax=self.ax, node_size=1500,
                             node_color=node_colors, alpha=0.8)

        labels = {n: f"{n}\n{y}" for n, y in G.nodes(data='y')}
        nx.draw_networkx_labels(G, self.pos, labels=labels, ax=self.ax, font_size=10)

        nx.draw_networkx_edges(G, self.pos, ax=self.ax, arrows=True,
                              edge_color='gray', width=2)

        self.ax.set_title(title, fontsize=14)
        plt.axis('off')
        display(self.fig)
        clear_output(wait=True)
        time.sleep(1)

class VisualTreap(Treap):
    def __init__(self):
        super().__init__()
        self.visualizer = TreapVisualizer()

    def split(self, node: Optional[Node], x: int) -> Tuple[Optional[Node], Optional[Node]]:
        if node is None:
            return None, None

        # Визуализация перед split
        self.visualizer.draw_treap(self, f"Before split at {x}", highlight=x)

        if node.x < x:
            left, right = super().split(node.right, x)
            node.right = left

            # Визуализация после split (правая часть)
            temp_treap = VisualTreap()
            temp_treap.root = node
            self.visualizer.draw_treap(temp_treap, f"Split right at {x}",
                                     highlight_nodes=[node.x, *(n.x for n in [left, right] if n)])

            return node, right
        else:
            left, right = super().split(node.left, x)
            node.left = right

            # Визуализация после split (левая часть)
            temp_treap = VisualTreap()
            temp_treap.root = node
            self.visualizer.draw_treap(temp_treap, f"Split left at {x}", highlight_nodes=[node.x, *(n.x for n in [left, right] if n)])

            return left, node

    def merge(self, left: Optional[Node], right: Optional[Node]) -> Optional[Node]:
        if left is None:
            return right
        if right is None:
            return left

        # Визуализация перед merge
        temp_treap = VisualTreap()
        temp_treap.root = left
        self.visualizer.draw_treap(temp_treap, "Before merge (left part)",
                                 highlight_nodes=[left.x, right.x])

        if left.y > right.y:
            left.right = super().merge(left.right, right)

            # Визуализация после merge (левый узел главный)
            temp_treap = VisualTreap()
            temp_treap.root = left
            self.visualizer.draw_treap(temp_treap, "After merge (left dominates)",
                                     highlight_nodes=[left.x, right.x])
            return left
        else:
            right.left = super().merge(left, right.left)

            # Визуализация после merge (правый узел главный)
            temp_treap = VisualTreap()
            temp_treap.root = right
            self.visualizer.draw_treap(temp_treap, "After merge (right dominates)",
                                     highlight_nodes=[left.x, right.x])
            return right

    def insert(self, x: int, y: Optional[float] = None):
        self.visualizer.draw_treap(self, f"Before inserting {x}")
        super().insert(x, y)
        self.visualizer.draw_treap(self, f"After inserting {x}", highlight=x)

    def remove(self, x: int):
        self.visualizer.draw_treap(self, f"Before removing {x}", highlight=x)
        super().remove(x)
        self.visualizer.draw_treap(self, f"After removing {x}")

# Демонстрация работы split и merge
def demo_split_merge():
    treap = VisualTreap()

    # Создаем дерево
    for x in [4, 2, 6, 1, 3, 5, 7]:
        treap.insert(x)
        time.sleep(1)

    # Демонстрация split
    left, right = treap.split(treap.root, 4)

    # Визуализация результатов split
    left_treap = VisualTreap()
    left_treap.root = left
    left_treap.visualizer.draw_treap(left_treap, "Left after split at 4")

    right_treap = VisualTreap()
    right_treap.root = right
    right_treap.visualizer.draw_treap(right_treap, "Right after split at 4")

    time.sleep(2)

    # Демонстрация merge
    merged = treap.merge(left, right)
    merged_treap = VisualTreap()
    merged_treap.root = merged
    merged_treap.visualizer.draw_treap(merged_treap, "After merge")

    plt.close()

# Запускаем демо
demo_split_merge()
```
